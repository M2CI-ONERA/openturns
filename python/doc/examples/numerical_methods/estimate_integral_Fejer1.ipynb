# Estimate an integral

In this basic example we are going to evaluate an integral of the form.

$$I_f = \int_{a}^{b}\, \int_{l_1(x_0)}^{u_1(x_0)}\, \int_{l_2(x_0, x_1)}^{u_2(x_0,x_1)}\, \int_{l_{n-1}(x_0, \dots, x_{n-2})}^{u_{n-1}(x_0, \dots, x_{n-2})} \, f(x_0, \dots, x_{n-1})\mathrm{d}{x_{n-1}}\dots\mathrm{d}{x_0}$$

with the iterated quadrature algorithm.

```python
from __future__ import print_function
import openturns as ot
import math as m
import numpy as np
```

```python
f = ot.SymbolicFunction(['x','y'], ['sin(x)+y'])
a = -2.5
b = 4.5
algo = ot.GaussLegendre([10,10])
value, nodes = algo.integrateWithNodes(f, ot.Interval([a,a], [b,b]))
print(value,nodes)
```

```python
f = ot.SymbolicFunction(['x','y'], ['sin(x)+y'])
a = -2.5
b = 4.5
algo = ot.Fejer1([10,10])
value, nodes = algo.integrateWithNodes(f, ot.Interval([a,a], [b,b]))
print(value,nodes)
```

```python
g = ot.Graph('Integration nodes', 'x', 'y', True, 'topright')
g.add(f.draw([a,a],[b,b]))
integration_nodes = ot.Cloud(nodes)
integration_nodes.setColor("magenta")
integration_nodes.setPointStyle('circle')
g.add(integration_nodes)
g
```
